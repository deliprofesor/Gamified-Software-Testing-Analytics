```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# 0. Setup: Create a directory for saving figures
output_dir = '../reports/figures'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Set aesthetic style
sns.set_theme(style="whitegrid")

# 1. Load the dataset
# Ensure the path is correct relative to your notebooks folder
file_path = '../data/gamification-software-testing-engagement-performance-test-effectiveness.feather'
df = pd.read_feather(file_path)

```

```python
# 2. Performance Distribution Across Exercises (Box Plot)
effectiveness_cols = [
    'Exercise1 Effectiveness', 
    'Exercise2 Effectiveness', 
    'Exercise3 Effectiveness', 
    'Exercise4 Effectiveness'
]

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[effectiveness_cols], palette="Set2")
plt.title('Distribution of Student Effectiveness Across Exercises', fontsize=15)
plt.ylabel('Effectiveness Score (0-100)', fontsize=12)
plt.xlabel('Exercise Type', fontsize=12)

# Save the figure
plt.savefig(f'{output_dir}/01_effectiveness_distribution.png', dpi=300, bbox_inches='tight')
plt.show()



```

```python
# 3. Engagement vs. Performance (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df, 
    x='Exercise2 Active Time', 
    y='Exercise2 Effectiveness', 
    hue='Academic Year',
    alpha=0.7,
    s=100
)

plt.title('Exercise 2: Impact of Active Time on Effectiveness', fontsize=15)
plt.xlabel('Active Time (Sessions/Minutes)', fontsize=12)
plt.ylabel('Effectiveness Score', fontsize=12)
plt.legend(title='Academic Year', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.savefig(f'{output_dir}/02_active_time_vs_effectiveness.png', dpi=300, bbox_inches='tight')
plt.show()


```

```python
# 4. Correlation Heatmap (Bonus: To see global trends)
plt.figure(figsize=(12, 10))
correlation_matrix = df[effectiveness_cols + ['Race Active Time', 'Race Number Executions']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Performance and Engagement Metrics', fontsize=15)

plt.savefig(f'{output_dir}/03_correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"Successfully saved all figures to: {output_dir}")
```
